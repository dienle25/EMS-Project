```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, LabelEncoder

# Tải dữ liệu
df = pd.read_csv('data.csv')

# Xem trước dữ liệu
display(df.head())

```

```python
# 1. Bỏ các dòng trùng lặp
df = df.drop_duplicates()

# 2. Xử lý missing values (Loại bỏ hoặc điền giá trị thay thế)
df = df.dropna()

# 3. Xử lý 45 sinh viên "High Effort, Low Score" (Tùy chỉnh điều kiện theo phân tích của nhóm)
# Ví dụ minh họa: df = df[~((df['study_hours'] > 8) & (df['exam_score'] < 40))]

# 4. Mã hóa các biến phân loại thành số
categorical_cols = ['gender', 'course', 'internet_access', 'sleep_quality', 'study_method', 'facility_rating', 'exam_difficulty']
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
```

```python
# Xác định các cột dạng số cần chuẩn hóa
# Lưu ý: Không chuẩn hóa cột mục tiêu 'exam_score' và cột định danh 'student_id'
numeric_features = ['age', 'study_hours', 'class_attendance', 'sleep_hours']

# Khởi tạo MinMaxScaler (đưa giá trị về khoảng 0 - 1)
scaler = MinMaxScaler()

# Áp dụng chuẩn hóa
df[numeric_features] = scaler.fit_transform(df[numeric_features])

# Kiểm tra lại sau khi chuẩn hóa
display(df.head())
```

```python
# Cột student_id không có tác dụng dự đoán điểm số, nên loại bỏ trước khi train mô hình
df_final = df.drop(columns=['student_id'])

# Xuất file CSV mới không chứa cột index
file_name = 'dataADY201m_cleaned_normalized.csv'
df_final.to_csv(file_name, index=False)

print(f"Đã lưu thành công file {file_name}!")
```
